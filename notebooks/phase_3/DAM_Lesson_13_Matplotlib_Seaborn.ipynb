```python
# Setup: Import libraries and load datasets
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Netflix data
df_netflix = pd.read_csv('https://raw.githubusercontent.com/zsmithc2c/dam-course/main/datasets/phase_3/netflix_titles.csv')
df_netflix['country'] = df_netflix['country'].fillna('Unknown')
df_netflix['director'] = df_netflix['director'].fillna('Not Listed')

# Load employee data for salary analysis
df_emp = pd.read_csv('https://raw.githubusercontent.com/zsmithc2c/dam-course/main/datasets/phase_1/employee_data.csv')

print(f"Netflix: {len(df_netflix)} titles")
print(f"Employee: {len(df_emp)} records")
```

## Exercise 1: Seaborn Themes

Seaborn instantly makes your charts look more professional with `sns.set_theme()`.

```python
# Exercise 1: Apply Seaborn themes

# TODO: Apply the default Seaborn theme
# Hint: sns.set_theme()

# Create a simple chart to see the theme
plt.figure(figsize=(8, 4))
df_netflix['type'].value_counts().plot(kind='bar')
plt.title('Netflix Content Types (with Seaborn theme)')
plt.tight_layout()
plt.show()

```

## Exercise 2: `sns.countplot()` vs `plt.bar()`

Seaborn's countplot automatically counts and plots categorical data — no need to call `value_counts()` first.

```python
# Exercise 2: Compare countplot to bar chart

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left: Matplotlib approach (manual)
type_counts = df_netflix['type'].value_counts()
axes[0].bar(type_counts.index, type_counts.values, color=['#E50914', '#221f1f'])
axes[0].set_title('Matplotlib: plt.bar()')
axes[0].set_ylabel('Count')

# Right: Seaborn approach (automatic)
# TODO: Use sns.countplot() on axes[1]
# Hint: sns.countplot(data=df_netflix, x='type', ax=axes[1])

axes[1].set_title('Seaborn: sns.countplot()')

plt.tight_layout()
plt.show()

```
