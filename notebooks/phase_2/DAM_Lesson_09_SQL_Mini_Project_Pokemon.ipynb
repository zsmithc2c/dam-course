```python
# Setup: Import libraries and load Pokemon data into SQLite
import sqlite3
import pandas as pd

# Load the Pokemon CSV
df = pd.read_csv('https://raw.githubusercontent.com/zsmithc2c/dam-course/main/datasets/phase_2/pokemon.csv')
print(f"Loaded {len(df)} Pokemon")

# Create an in-memory SQLite database and load the data
conn = sqlite3.connect(':memory:')
df.to_sql('pokemon', conn, if_exists='replace', index=False)

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql(query, conn)

# Preview the data
run_sql('SELECT * FROM pokemon LIMIT 5')
```

```python
# Challenge 6: Average HP by type
# YOUR SQL QUERY HERE
query_6 = """

"""
run_sql(query_6)

# ANSWER: SELECT type1, ROUND(AVG(hp), 1) AS avg_hp FROM pokemon GROUP BY type1 ORDER BY avg_hp DESC
```

### Challenge 7: Types with High Average Attack

```python
# Challenge 7: Types with high average attack
# YOUR SQL QUERY HERE
query_7 = """

"""
run_sql(query_7)

# ANSWER: SELECT type1, ROUND(AVG(attack), 1) AS avg_attack FROM pokemon GROUP BY type1 HAVING AVG(attack) > 80 ORDER BY avg_attack DESC
```
